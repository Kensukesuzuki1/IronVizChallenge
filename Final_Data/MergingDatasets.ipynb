```python
import pandas as pd
```

```python
years = [str(2000+x) for x in range(13,19)]
```

```python
cod_df = pd.DataFrame()
for year in years:
    csv_path = f"{year}_Suicide_Data.csv"
    df = pd.read_csv(csv_path, dtype=object)
    cod_df = pd.concat([cod_df, df])
                       
cod_df
```

```python
cod_df.to_csv("Merged_Suicide_Data.csv") 
```
