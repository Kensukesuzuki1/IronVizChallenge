```python
df2["County"] = list(map(lambda i: i[ : -4], df2["County"])) 
df2["County"]
```

```python
CountyCleanup = {' County', ' Borough', " Census Area", " Parish"}
```

```python
p = re.compile('|'.join(map(re.escape, CountyCleanup)))
df2["County"] = [p.sub('', s) for s in df2["County"]]
```
